```python


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_excel('file.xlsx',parse_dates=['Date'],index_col="Date")


#Month_wise Debit
dfd=df[df['Nature']=="Debit"]




arr_avg=np.array(dfd['Amount'].resample("M").mean())
arr_max=np.array(dfd['Amount'].resample("M").max())
arr_min=np.array(dfd['Amount'].resample("M").min())

x_value=np.arange(1,13)



inp_one=int(input("Debit Statement in Month-Wise:\n1.Maximum Rate\n2.Average rate\n3.Maximum Rate"))

if inp_one==2:
    plt.plot(x_value,arr_avg,linewidth=3,color='orange',label='Average Debit Rate',marker='o',markersize=5,markeredgecolor='black')


elif inp_one==1:
    plt.plot(x_value,arr_max,linewidth=3,color='red',label='Maximum Debit Rate',marker='o',markersize=5,markeredgecolor='black')

elif inp_one==3:
    plt.plot(x_value,arr_min,linewidth=3,color='green',label='Minimum Debit Rate',marker='o',markersize=5,markeredgecolor='black')


"""y=dfd['Amount'].resample("M").mean()
max=dfd['Amount'].resample("M").max()
min=dfd['Amount'].resample("M").min()

x=np.arange(1,12)

plt.plot(x=np.arange(1,12),y=dfd['Amount'].resample("M").mean())"""
```
